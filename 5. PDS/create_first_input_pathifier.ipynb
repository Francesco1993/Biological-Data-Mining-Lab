### EXTRACTION of PDS from each Pathway

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle("../2. batch-effect-evaluation/data/merged_df_no_batch_effect.pkl")
df.head()
```

```python
a = list(df.columns)
pathifier_index = []
for el in a:
    if "L" in el:
        pathifier_index.append("1")
    else:
        pathifier_index.append("0")

path_index = np.array([pathifier_index])
path_index_df = pd.DataFrame(path_index)
path_index_df.columns = df.columns

concat_path_index = pd.concat([path_index_df, df])
concat_path_index.index = ['NORMALS'] + list(df.index)

concat_path_index.head()
```

```python
concat_path_index.to_csv('data/first_input_pathifier.csv', sep='\t')
```
